# Facets

## Splitting Data into Subplots with Facets

Use **facet_grid() **and specify the variables on which to split.

```python
p = ggplot(mpg, aes(x='displ', y='hwy')) + geom_point()
```

```python
# arranged by row
p +coord_fixed(0.05)+ facet_grid('drv~.')
```

```python
# arranged by column
p +coord_fixed(0.2)+ facet_grid('.~cyl')
```

```python
(p
 + facet_grid('drv~cyl')
)
```

With** facet_wrap()**, the subplots are laid out **horizontally** and wrap around, like words on a page  
failed in 0.2.1

```python
mpg=mpg.rename(columns = {'class':'cls'})
```

```python
(ggplot(mpg, aes(x='displ', y='hwy')) + geom_point() 
 + facet_wrap('~cls',ncol=4)
) # can not use keywords in strings 0.2.1
```

## Using Facets with Different Axes

subplots with different ranges or items on their axes.  
Set the scales to "free_x", "free_y", or "free"

```python
# With free y scales
p + facet_grid('drv ~ cyl', scales="free_y")
```

```python
# With free x and y scales
p + facet_grid('drv ~ cyl', scales="free")
```

```python
mpg2 = mpg.copy()
mpg2['drv']=mpg['drv'].replace({'4':'$4^{wd}$','f':'$Front \  \pi $','r':'Rear'})
```

```python
ggplot(mpg2, aes(x='displ', y='hwy')) + geom_point() + facet_grid('drv ~ .')
```

## Changing the Appearance of Facet Labels and Headers

set **strip_text** to control the text appearance  
**strip_background** to control the background appearance

```python
(ggplot(cabbage_exp, aes(x='Cultivar', y='Weight')) 
 + geom_bar(stat="identity") 
 +facet_grid('. ~ Date') 
 +theme(strip_text = element_text(face="bold", size=14,color='black'),
strip_background=element_rect(fill="lightblue", colour="black",size=1))
)
```
