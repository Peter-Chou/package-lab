# Legends

## Remove the legend for fill

```python
p = (ggplot(PlantGrowth, aes(x='group', y='weight', fill='group'))
 + geom_boxplot())
p + scale_fill_discrete(guide=False) # 
```

## Changing the Position of a Legend

```python
p + theme(legend_position=(0.82,0.25),
         legend_background = element_rect(fill='orange',color='black')) # 'top','bottom' also works
```

## Reversing the Order of Items in a Legend

```python
p + guides(fill=guide_legend(reverse=True))
```

## Changing a Legend Title

```python
(p + labs(fill='Condition') 
 + theme(legend_title=element_text(face='italic',family='times',color='red',size=14))
)
```

```python
hw = (ggplot(heightweight, aes(x='ageYear', y='heightIn', color='sex')) +
      geom_point(aes(size='weightLb')) + scale_size_continuous(range=[1,4])
     )
hw
```

```python
hw + labs(color='Male/Female',size='Weight (pounds)')
```

## Removing a Legend Title

```python
p + guides(fill = guide_legend(title=''))
```

## Changing the Labels in a Legend

```python
p + scale_fill_discrete(labels=['Control','Treatment 1','Treatment 2'])
```

```python
(hw + labs(color='Male/Female',size='Weight (pounds)')
 + scale_color_discrete(labels=['Female','Male'])
)
```

## Changing the Appearance of Legend Labels

```python
(p + theme(legend_text=element_text(face="italic", family="Times", colour="red",
size=14)))
```
