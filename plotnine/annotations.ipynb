## Adding Error Bars

Use geom_errorbar and map variables to the values for ymin and ymax. Adding the error
bars is done the same way for bar graphs and line graphs

```python
cabbage_exp.query('Cultivar=="c39"')
```

```python
(ggplot(cabbage_exp.query('Cultivar=="c39"'),aes('Date','Weight')) +
 geom_bar(fill='white',color='black',stat='identity') +
 geom_errorbar(aes(ymin='Weight-se',ymax='Weight+se'),width=0.15)
)
```

```python
(ggplot(cabbage_exp.query('Cultivar=="c39"'),aes('Date','Weight')) +
 geom_line(aes(group=1)) +
 geom_point(size=4) +
 geom_errorbar(aes(ymin='Weight-se',ymax='Weight+se'),width=0.2)
)
```

 position="dodge", which is shorthand for position=position_dodge()

```python
(ggplot(cabbage_exp, aes(x='Date', y='Weight', fill='Cultivar')) +
 geom_bar(position="dodge",stat='identity') +
 geom_errorbar(aes(ymin='Weight-se', ymax='Weight+se'),position=position_dodge(0.9), width=.2)
)
```

```python
(ggplot(cabbage_exp, aes(x='Date', y='Weight', colour='Cultivar', group='Cultivar')) +
geom_errorbar(aes(ymin='Weight-se', ymax='Weight+se'),
width=.2, size=0.25, colour="black", position=position_dodge(.15)) +
geom_line(position=position_dodge(.15)) +
geom_point(position=position_dodge(.15), size=2.5))
```

## Adding Annotations to Individual Facets

A data frame with labels for each facet

```python
f_labels = pd.DataFrame({'drv':['4','f','r'],'label':['4wd','Front','Rear']})
```

```python
f_labels
```

```python
(ggplot(mpg,aes('displ','hwy')) +
 geom_point() +
 facet_grid('. ~ drv') +
 geom_text(data=f_labels,label=f_labels['label'],x=5.5, y=40,size=14) # dataframe works like an anchor
)
```

If use annotate(), the label will appear in all facets

```python
(ggplot(mpg,aes('displ','hwy')) +
 geom_point() +
 facet_grid('. ~ drv') +
 geom_text(data=f_labels,label='label test',x=5, y=40,size=14) # dataframe works like an anchor
)
```

```python
label_for_r = pd.DataFrame({'drv':['4','f','r'],
 'formula':[r'$y = 30.68 - 2.88x$',r'$y = 37.38 - 3.6x$',r'$y = 25.78 - 0.92x$'],
 'r2':[r'$R^2 = 0.65$',r'$R^2 = 0.36$',r'$R^2 = 0.04$']
})
```
